```python
df=pd.read_excel (r'C:\Users\HP\Downloads\data AI\dataset\dataset.xlsx')
df
```

```python
df["words"] = df["PLACE"] +" "+ df["TITLE"] +" "+ df["SKILLS"] +" "+ df["DEGREE"] +" "+ df["EDUCATION"] +" "+ df["POST"] +" "+ df["RESPONSIBILITY"]
```

```python
df['PLACE'].str.replace('[^\w\s]','',regex=True)
df['PLACE']=df['PLACE'].str.lower()
df['POST'].str.replace('[^\w\s]','',regex=True)
df['POST']=df['POST'].str.lower()
df['SKILLS'].str.replace('[^\w\s]','',regex=True)
df['SKILLS']=df['SKILLS'].str.lower()
df['DEGREE'].str.replace('[^\w\s]','',regex=True)
df['DEGREE']=df['DEGREE'].str.lower()
df['words'].str.replace('[^\w\s]','',regex=True)
df['words']=df['words'].str.lower()
```

```python
stop = set(stopwords.words('english'))
pat = r'\b(?:{})\b'.format('|'.join(stop))
df['PLACE']=df['PLACE'].str.replace(pat, ' ',regex=True)
df['words']=df['words'].str.replace(pat, ' ',regex=True)
df['SKILLS']=df['SKILLS'].str.replace(pat, ' ',regex=True)
df['DEGREE']=df['DEGREE'].str.replace(pat, ' ',regex=True)
df['POST']=df['POST'].str.replace(pat, ' ',regex=True)
```

```python
import string
def remove_punctuations(df):
    for punctuation in string.punctuation:
        df = df.replace(punctuation, '')
    return df
df["PLACE"] = df['PLACE'].apply(remove_punctuations)
df["SKILLS"] = df['SKILLS'].apply(remove_punctuations)
df["POST"] = df['POST'].apply(remove_punctuations)
df["DEGREE"] = df['DEGREE'].apply(remove_punctuations)
df["words"] = df['words'].apply(remove_punctuations)
```
